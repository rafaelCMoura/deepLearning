### Importando a biblioteca

```python
import tensorflow as tf
```

### Criando um modelo (composto de variáveis e equações)

```python
x = tf.constant([35, 40, 45], name='x')
y = tf.Variable(x + 5, name='y')
print(y)
```

Veja que ao imprimir o resultado de y, não aparece o resultado da operação. Apenas a estrutura da operação foi criada. Isso facilita o reuso de código.

### Executando o modelo criado

```python
model = tf.global_variables_initializer()

with tf.Session() as session:
    session.run(model)
    print(session.run(y))
```

### Variando os valores de um parêmtro no modelo

```python
x = tf.Variable(0, name='x')

model = tf.global_variables_initializer()

with tf.Session() as session:
    session.run(model)
    for i in range(5):
        x = x + 1
        print(session.run(x))
```

### Trabalhando com imagens

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

# Load the image
image = mpimg.imread("MarshOrchid.jpg")

plt.imshow(image)
plt.show()
```

### Manipulando a imagem com a tensorflow

```python
# Create a TensorFlow Variable
x = tf.Variable(image, name='x')

model = tf.global_variables_initializer()

with tf.Session() as session:
    x = tf.transpose(x, perm=[1, 0, 2])
    session.run(model)
    result = session.run(x)

plt.imshow(result)
plt.show()
```

### Usando placeholders

Placeholders são estruturas mais simples que são utilizadas para dados externos ao modelo (São o ponto de entrada, onde o modelo é alimentado). Um placeholder pode ter tamanho variável, uma variável não.

Geralmente se usam placeholders para dados externos e variáveis para dados que mudam de estado ao longo das operações no modelo. Em resumo temos:
- Variables e Constants (Para usar ao longo do modelo)
- Placeholders (Dados externos ao modelo, são o ponto de entrada de dados. ex: minibatches)

```python
x = tf.placeholder("float", None)
y = x * 2

with tf.Session() as session:
    result = session.run(y, feed_dict={x: [1, 2, 3]})
    print(result)
```
